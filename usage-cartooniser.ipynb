```python
import codeprojectai.core as cpai
from PIL import Image, ImageDraw
import pprint
import matplotlib.pyplot as plt
import base64
%matplotlib inline

IP = 'localhost'
PORT = '32168'
TIMEOUT = 20 # Default is 10
MIN_CONFIDENCE = 0.01 # Default is 0.45
CUSTOM_MODEL = "cartooniser"

cpaiobject = cpai.CodeProjectAICartoonise(ip=IP, port=PORT, timeout=TIMEOUT, min_confidence=MIN_CONFIDENCE)

image_path = 'tests/images/idris.jpg'
im = Image.open(image_path)
plt.imshow(im)

try:
    with open(image_path, 'rb') as image_bytes:
        response = cpaiobject.cartoonise(image_bytes)
        print(response)
        plt.imshow(base64.decodestring(response['imageBase64']));
except cpai.CodeProjectAIException as exc:
    print(exc)
```
