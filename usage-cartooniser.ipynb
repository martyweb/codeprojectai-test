```python
import codeprojectai.core as cpai
from PIL import Image, ImageDraw
from io import BytesIO
#import pprint
import matplotlib.pyplot as plt
import base64

IP = 'localhost'
PORT = '32168'
TIMEOUT = 20 # Default is 10


cpaiobject = cpai.CodeProjectAICartoonise(ip=IP, port=PORT, timeout=TIMEOUT)

image_path = 'tests/images/idris.jpg'
im = Image.open(image_path)
plt.imshow(im)

```

```python
try:
    with open(image_path, 'rb') as image_bytes:
        response = cpaiobject.cartoonise(image_bytes)
        #print(response)
        im = Image.open(BytesIO(base64.b64decode(response['imageBase64'])))
        plt.imshow(im);
except cpai.CodeProjectAIException as exc:
    print(exc)
```
